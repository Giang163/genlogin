```python
import pandas as pd
data = pd.read_csv("D:\download\MSN - MSN.csv")
data.head()
```

```python
import matplotlib.pyplot as plt

# Biểu đồ dạng nến
plt.figure(figsize=(12, 6))
plt.plot(data['Date/Time'], data['Close'], label='Giá đóng cửa')
plt.xlabel('Thời gian')
plt.ylabel('Giá cổ phiếu')
plt.title('Biểu đồ Giá cổ phiếu theo thời gian')
plt.legend()
plt.show()

```
