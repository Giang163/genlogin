```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_distribution(data, title):
    plt.figure(figsize=(10, 6))
    sns.histplot(data, kde=True, bins=30, color='blue', stat="density", linewidth=0)
    plt.title(title)
    plt.xlabel('Average Change Rate')
    plt.ylabel('Density')
    plt.grid(True)  # Thêm lưới cho dễ nhìn
    plt.show()

```

```python
def process_files_and_plot(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith('.xlsx'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_excel(file_path)
            if 'average_change_rate' in df.columns:
                plot_distribution(df['average_change_rate'], f'Distribution of Average Change Rate - {filename}')
            else:
                print(f"Column 'average_change_rate' not found in {filename}")
```

```python

def main():
    folder_path = 'productCode_files' 
    process_files_and_plot(folder_path)

if __name__ == '__main__':
    main()
```
