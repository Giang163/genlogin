```python
# Chọn các cột cần thiết từ DataFrame df
selected_columns = ['productId', 'productCode', 'productName', 'categoryId', 'categoryName', 'quantity',
                 'subTotal', 'returnQuantity', 'code', 'purchaseDate', 'customerCode', 'statusValue' ]  # Thay thế bằng tên các cột bạn muốn chọn

# Tạo một DataFrame mới chỉ chứa các cột đã chọn
df = df[selected_columns].copy()

# In ra dữ liệu đầu tiên trong DataFrame mới
df
```

```python
product_sales = df.groupby('productCode').agg({'quantity': 'sum', 'productName': 'first'})

product_sales_sorted = product_sales.sort_values(by='quantity', ascending=False)

associated_products = {}

for product_A_id, _ in product_sales_sorted.iterrows():

    customers_with_A = df[df['productCode'] == product_A_id]['customerCode'].unique()

    other_products = df[(df['customerCode'].isin(customers_with_A)) & (df['productCode'] != product_A_id)].groupby('productCode')['quantity'].sum().reset_index()
    other_products = other_products.sort_values(by='quantity', ascending=False)
    associated_products[product_A_id] = other_products.head(20)

for product_A_id, associated_products_df in associated_products.items():
    product_A_name = product_sales.loc[product_A_id, 'productName']
    print(f"Products associated with '{product_A_name}' (Product ID: {product_A_id}):")
    
    for i, (_, row) in enumerate(associated_products_df.iterrows(), start=1):
        product_B_id = row['productCode']
        product_B_name = product_sales.loc[product_B_id, 'productName']
        quantity = row['quantity']
        print(f"{i}. {product_B_name} (productCode: {product_B_id}), quantity: {quantity}")
    print()
```

```python
import pandas as pd
from tabulate import tabulate

# Đầu vào là DataFrame có 2 cột là 'productName' và 'customerCode'
# df là DataFrame chứa dữ liệu của bạn

# Bước 1: Nhóm dữ liệu theo cặp sản phẩm và tính số lượng khách hàng cho mỗi cặp sản phẩm
grouped = df.groupby(['productName', 'customerCode']).size().reset_index(name='count')

# Bước 2: Tìm cặp sản phẩm (A và B) có tổng số khách hàng mua cả hai sản phẩm là lớn nhất với mỗi sản phẩm A
def find_product_b_with_max_customers_for_both(df):
    product_names = df['productName'].unique()
    best_product_pairs = []

    for product_a in product_names:
        max_customers = 0
        best_product_b = None

        for product_b in product_names:
            if product_a != product_b:
                customers_for_both = set(df[(df['productName'] == product_b) & (df['customerCode'].isin(set(df[df['productName'] == product_a]['customerCode'])))]
                                          ['customerCode'])
                customers_count = len(customers_for_both)

                if customers_count > max_customers:
                    max_customers = customers_count
                    best_product_b = product_b

        if best_product_b is not None:
            best_product_pairs.append((product_a, best_product_b, max_customers))

    return best_product_pairs

# Tìm các cặp sản phẩm có tổng số khách hàng mua cả hai sản phẩm là lớn nhất với mỗi sản phẩm A
best_product_pairs = find_product_b_with_max_customers_for_both(df)

# Chuyển kết quả thành DataFrame để sử dụng tabulate
result_df = pd.DataFrame(best_product_pairs, columns=['Product A', 'Product B', 'Số lượt khách hàng'])

# In kết quả dưới dạng bảng bằng tabulate
print(tabulate(result_df, headers='keys', tablefmt='psql'))
```
