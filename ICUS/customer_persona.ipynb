```python
import pandas as pd
```

```python
df = pd.read_excel("DanhSachKhachHang_segmented (1)_icus.xlsx")
df.head(5)
```

```python
import matplotlib.pyplot as plt

transaction_freq = df['Số lần giao dịch'].dropna()

# Vẽ histogram
plt.figure(figsize=(10, 6))
plt.hist(transaction_freq, bins=50, color='skyblue', edgecolor='black')
plt.title('Phân bố Tần suất Giao dịch của Khách Hàng')
plt.xlabel('Số Lần Giao Dịch')
plt.ylabel('Số Lượng Khách Hàng')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
