```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("final_result1.xlsx")
```

```python
from datetime import datetime

# Chuyển đổi cột purchaseDate sang định dạng datetime
df['purchaseDate'] = pd.to_datetime(df['purchaseDate'])

# Tính toán số ngày chênh lệch đến thời điểm hiện tại
current_date = datetime.now()
df['daysDifference'] = (current_date - df['purchaseDate']).dt.days
```

```python
grouped_df = df.drop_duplicates(subset=['customerCode', 'productName', 'daysDifference'])

# Chọn chỉ các cột quan trọng để đưa vào bảng mới
result_df = grouped_df[['customerCode','productCode', 'productName', 'daysDifference']]

# Hiển thị kết quả
result_df
```

```python
grouped2  = result_df.groupby(['customerCode','productCode', 'productName'])['daysDifference'].agg(['max', 'min', 'count']).reset_index()
grouped2 .columns = ['customerCode', 'productCode','productName', 'max_daysDifference', 'min_daysDifference', 'count_daysDifference']

# Tính toán công thức
grouped2 ['average_change_rate'] = (grouped2 ['max_daysDifference'] - grouped2 ['min_daysDifference']) / (grouped2 ['count_daysDifference'] - 1)
```

```python
grouped2
```

```python
grouped3 = grouped2[['customerCode', 'productCode', 'productName', 'average_change_rate']]
grouped3 = grouped3.dropna(subset = ['average_change_rate'])
```

```python
grouped3
```

```python
output_file = 'output.xlsx'

# Tạo một đối tượng ExcelWriter
with pd.ExcelWriter(output_file, engine='openpyxl', mode='w') as writer:
    # Ghi DataFrame 'grouped' vào sheet 'Results'
    grouped3.to_excel(writer, sheet_name='Results', index=False)

print(f"Kết quả đã được lưu vào file '{output_file}'")
```

```python
# tách file theo productCode
import os

# Nhóm dữ liệu theo 'productCode'
grouped4 = grouped3.groupby('productCode')

# Đường dẫn thư mục để lưu các file
output_folder = 'productCode_files'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Lặp qua từng nhóm và lưu vào file Excel riêng
for product_code, data in grouped4:
    output_file = os.path.join(output_folder, f'{product_code}.xlsx')
    data.to_excel(output_file, index=False)
```
